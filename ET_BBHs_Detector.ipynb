#### Below we used tow data samples, each of which contains data of one hour and 15 seconds length. Sources in the tow samples belong to 4-5 and >8 SNR range.
#### The model used below is ResNet101, and it  was trained on sources belong to 5 SNR ranges: 4-5, 5-6, 6-7, 7-8 and > 8.

```python
from ET_Processings import load_classifier, ET_BBH_detector
import numpy as np
import pandas as pd
import shutil
import os
```

```python
model_path = "ResNet.h5" # Path to the trained deepl learning model
detector = load_classifier(model_path) 
ts_sample_4_5 = "snr_4_723_et_sim.txt" # A data sample contians sources with SNR of 4 to 5
ts_sample_g8 = "723_8_inj_et_sim.txt" # A data sample contians sources with SNR of 8 or higher

file = 'detection_report.txt'
if os.path.exists(file):
    os.remove(file) 
```

### Sources with SNR > 8:

```python
%%time
ET_BBH_detector(ts_sample_g8, detector,file)
```
